Находим и удаляем дубликаты

```python
dupl_columns = list(df_cafe_cleaned.columns)

mask = df_cafe_cleaned.duplicated(subset=dupl_columns)
cafe_duplicates = df_cafe_cleaned[mask]
print(f'Number of dublicates: {cafe_duplicates.shape[0]}')
```

```python
df_cafe_cleaned.drop_duplicates(inplace=True)
```

Выявляем признаки с большим количеством уникальных и одинаковых значений

```python
low_information_cols = []

#cycle through all columns
for col in df_cafe_cleaned.columns:
    #the highest relative frequency in the feature
    top_freq = df_cafe_cleaned[col].value_counts(normalize=True).max()
    #the proportion of unique values from the size of the feature
    nunique_ratio = df_cafe_cleaned[col].nunique() / df_cafe_cleaned[col].count()
    # compare the highest frequency with the threshold
    if top_freq > 0.7:
        low_information_cols.append(col)
        print(f'{col}: {round(top_freq*100, 2)}% identical values')
    # compare the unique frequency with the threshold
    if nunique_ratio > 0.7:
        low_information_cols.append(col)
        print(f'{col}: {round(nunique_ratio*100, 2)}% unique values')
len(low_information_cols)
```

```python
# убедиться, что колонка — date
df_cafe_cleaned['Дата'] = pd.to_datetime(df_cafe_cleaned['Время открытия'], errors='coerce').dt.date
```

```python
ru_holidays = holidays.Russia(years=range(2021, 2026), language='ru')
print(ru_holidays['2021-01-01'])  # -> 'Новый год'
```

```python
# Создаем объект с праздниками для России
ru_holidays = holidays.Russia(years=range(2021, 2026), language='ru')
```

```python
# получить имя праздника (None для обычных дней) и флаг
df_cafe_cleaned['holiday_name'] = df_cafe_cleaned['Дата'].map(lambda d: ru_holidays.get(d))
df_cafe_cleaned['is_holiday'] = df_cafe_cleaned['holiday_name'].apply(lambda x: 0 if pd.isna(x) else 1)
```

```python
df_cafe_cleaned['holiday_name'].value_counts(dropna=False).head(30)
```

```python
# заменить NaN на текст
df_cafe_cleaned['holiday_name'] = df_cafe_cleaned['holiday_name'].fillna('Рабочий день')

# унифицировать перенесённые выходные в простое "Выходной"
df_cafe_cleaned['holiday_name'] = df_cafe_cleaned['holiday_name'].astype(str).str.replace(
    r'Выходнойs*(.*?)', 'Выходной', regex=True)

# объединить разные варианты Нового года в одно название
df_cafe_cleaned['holiday_name'] = df_cafe_cleaned['holiday_name'].replace({
    'Новый год': 'Новогодние каникулы',
    'Новогодние каникулы': 'Новогодние каникулы'
})
```
